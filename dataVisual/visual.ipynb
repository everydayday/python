```python

from matplotlib import pyplot as plt

x1 = list(range(5))
y1 = [i ** 2 for i in x1]

x2 = list(range(5))
y2 = [i ** 3 for i in x2]

fig, ax = plt.subplots()
ax.plot(x1, y1, label = "AAA", color="red",linewidth = 5)
ax.scatter(x1,y1, color = "red")
ax.plot(x2, y2, label = "BBB", color = "black")
ax.scatter(x2,y2, color = "black")
ax.set_title('AAA vs BBB')
ax.set_xlabel('x')
ax.set_ylabel('y')

ax.set_major_xxx

plt.show()
```
